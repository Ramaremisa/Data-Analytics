### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(df.columns)
```

```python
# Check what you actually have:
print(df['SOPartFreq'].unique())
```

```python
freq_order = [
    'Less than once per month or never',
    'A few times per month',
    'A few times per week',
    'Daily or almost daily'
]

freq_mapping = {label: i for i, label in enumerate(freq_order, start=1)}

df['SOPartFreq_num'] = df['SOPartFreq'].map(freq_mapping)
```

```python
##Write your code here
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# Map the "Age" column to numeric values
df['Age_numeric'] = df['Age'].map(age_mapping)
```

```python
# Group by age and calculate frequency and mean SOPartFreq
bubble_df = df.groupby('Age_numeric').agg(
    freq=('SOPartFreq_num', 'count'),
    avg_SO=('SOPartFreq_num', 'mean')
).reset_index()

plt.figure(figsize=(12, 7))

plt.scatter(
    bubble_df['Age_numeric'],      # X-axis → mapped age values
    bubble_df['avg_SO'],           # Y-axis → mean SOPartFreq
    s=bubble_df['freq'] * 20,      # Bubble size → frequency
    alpha=0.6
)

plt.title("Bubble Plot: Age vs SOPartFreq")
plt.xlabel("Age (numeric midpoint)")
plt.ylabel("Average SOPartFreq")
plt.grid(True)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# 1. Age mapping (you already have this)
# -----------------------------
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# Create numeric age column (safe re-run)
df['Age_numeric'] = df['Age'].map(age_mapping)

# -----------------------------
# 2. Prepare data for plotting
# -----------------------------
# Keep only needed columns and drop missing values
plot_cols = ['ConvertedCompYearly', 'JobSat', 'Age_numeric']
df_plot = df[plot_cols].dropna()

# (Optional) remove zero/negative compensation if they exist
df_plot = df_plot[df_plot['ConvertedCompYearly'] > 0]

# Map JobSat (text) → numeric codes for y-axis
job_cat = df_plot['JobSat'].astype('category')
df_plot['JobSat_code'] = job_cat.cat.codes
job_labels = list(job_cat.cat.categories)

# -----------------------------
# 3. Bubble plot
# -----------------------------
plt.figure(figsize=(12, 7))

plt.scatter(
    df_plot['ConvertedCompYearly'],   # X: yearly compensation
    df_plot['JobSat_code'],          # Y: numeric job sat codes
    s=df_plot['Age_numeric'] * 5,    # Bubble size: age (scaled)
    alpha=0.5
)

# Replace y-ticks with the actual JobSat labels
plt.yticks(
    ticks=range(len(job_labels)),
    labels=job_labels
)

plt.xlabel("Yearly Compensation (ConvertedCompYearly)")
plt.ylabel("Job Satisfaction (JobSat)")
plt.title("Bubble Plot: Compensation vs Job Satisfaction\n(Bubble size represents Age)")
plt.grid(True, axis='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here

# Keep only needed rows
tmp = df[['Age_numeric', 'LanguageHaveWorkedWith']].dropna()

# Split multi-select languages and explode to long format
tmp = tmp.assign(Language=tmp['LanguageHaveWorkedWith'].str.split(';'))
tmp = tmp.explode('Language')

# Count respondents per (Age, Language)
lang_age = (
    tmp.groupby(['Age_numeric', 'Language'])
       .size()
       .reset_index(name='count')
)

# Encode language as categorical codes for y-axis
lang_age['Lang_code'] = lang_age['Language'].astype('category').cat.codes
lang_labels = lang_age['Language'].astype('category').cat.categories

plt.figure(figsize=(12, 7))
plt.scatter(
    lang_age['Age_numeric'],
    lang_age['Lang_code'],
    s=lang_age['count'] * 10,   # scale size
    alpha=0.6
)

plt.yticks(
    ticks=range(len(lang_labels)),
    labels=lang_labels
)

plt.xlabel("Age (numeric midpoint)")
plt.ylabel("Programming Language")
plt.title("Bubble Plot: Technology Preferences by Age\n(Bubble size = number of respondents)")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
tmp = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()

# Split and explode databases
tmp = tmp.assign(Database=tmp['DatabaseWantToWorkWith'].str.split(';'))
tmp = tmp.explode('Database')

# Count respondents per (Database, JobSat)
db_jobsat = (
    tmp.groupby(['Database', 'JobSat'])
       .size()
       .reset_index(name='count')
)

# Encode JobSat for y-axis
db_jobsat['JobSat_code'] = db_jobsat['JobSat'].astype('category').cat.codes
jobsat_labels = db_jobsat['JobSat'].astype('category').cat.categories

plt.figure(figsize=(12, 7))
plt.scatter(
    db_jobsat['Database'],
    db_jobsat['JobSat_code'],
    s=db_jobsat['count'] * 20,
    alpha=0.6
)

plt.xticks(rotation=45, ha='right')
plt.yticks(
    ticks=range(len(jobsat_labels)),
    labels=jobsat_labels
)

plt.xlabel("Preferred Database")
plt.ylabel("Job Satisfaction")
plt.title("Bubble Plot: Preferred Databases vs Job Satisfaction\n(Bubble size = number of respondents)")
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
import matplotlib.pyplot as plt
import pandas as pd

# ---------------------------------------------------
# 1. Prepare Data
# ---------------------------------------------------

# Map JobSat text → numeric values
job_sat_mapping = {
    'Very dissatisfied': 1,
    'Slightly dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Slightly satisfied': 4,
    'Very satisfied': 5
}

tmp = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()

# Remove zero/negative salaries
tmp = tmp[tmp['ConvertedCompYearly'] > 0]

# Convert JobSat to numeric scale
tmp['JobSat_num'] = tmp['JobSat'].map(job_sat_mapping)

# Split DevType (multi-select) and explode
tmp = tmp.assign(Role=tmp['DevType'].str.split(';'))
tmp = tmp.explode('Role')

# Aggregate: average compensation & average job satisfaction per Role
role_stats = (
    tmp.groupby('Role')
       .agg(
            avg_comp=('ConvertedCompYearly', 'mean'),
            avg_jobsat=('JobSat_num', 'mean'),
            count=('JobSat_num', 'size')
       )
       .reset_index()
)

# Encode Role for x-axis
role_stats['Role_code'] = role_stats['Role'].astype('category').cat.codes
role_labels = role_stats['Role'].astype('category').cat.categories

# ---------------------------------------------------
# 2. Plot Bubble Chart
# ---------------------------------------------------
plt.figure(figsize=(14, 7))

plt.scatter(
    role_stats['Role_code'],           # X-axis: developer role
    role_stats['avg_comp'],            # Y-axis: average compensation
    s=role_stats['avg_jobsat'] * 300,  # Bubble size = average job satisfaction
    alpha=0.6
)

plt.xticks(
    ticks=range(len(role_labels)),
    labels=role_labels,
    rotation=45,
    ha='right'
)

plt.ylabel("Average Yearly Compensation (ConvertedCompYearly)")
plt.xlabel("Developer Role (DevType)")
plt.title("Bubble Plot: Compensation Across Developer Roles\n(Bubble size = Job Satisfaction)")
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
tmp = df[['Age_numeric', 'NEWCollabToolsHaveWorkedWith']].dropna()

# Split and explode tools
tmp = tmp.assign(Tool=tmp['NEWCollabToolsHaveWorkedWith'].str.split(';'))
tmp = tmp.explode('Tool')

# Count respondents per (Age, Tool)
tools_age = (
    tmp.groupby(['Age_numeric', 'Tool'])
       .size()
       .reset_index(name='count')
)

# Encode tools for y-axis
tools_age['Tool_code'] = tools_age['Tool'].astype('category').cat.codes
tool_labels = tools_age['Tool'].astype('category').cat.categories

plt.figure(figsize=(12, 7))
plt.scatter(
    tools_age['Age_numeric'],
    tools_age['Tool_code'],
    s=tools_age['count'] * 15,
    alpha=0.6
)

plt.yticks(
    ticks=range(len(tool_labels)),
    labels=tool_labels
)

plt.xlabel("Age (numeric midpoint)")
plt.ylabel("Collaboration Tool")
plt.title("Bubble Plot: Collaboration Tools by Age\n(Bubble size = frequency of tool use)")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here

# Keep only needed columns and drop missing
tmp = df[['WebframeWantToWorkWith', 'JobSat']].dropna()

# Split multi-select web frameworks and explode
tmp = tmp.assign(Webframe=tmp['WebframeWantToWorkWith'].str.split(';'))
tmp = tmp.explode('Webframe')

# Count respondents for each (Webframe, JobSat) pair
wf_js = (
    tmp.groupby(['Webframe', 'JobSat'])
       .size()
       .reset_index(name='count')
)

# Encode JobSat for y-axis
wf_js['JobSat_code'] = wf_js['JobSat'].astype('category').cat.codes
jobsat_labels = wf_js['JobSat'].astype('category').cat.categories

plt.figure(figsize=(12, 7))
plt.scatter(
    wf_js['Webframe'],          # x: web framework
    wf_js['JobSat_code'],       # y: job satisfaction (coded)
    s=wf_js['count'] * 20,      # bubble size: number of respondents
    alpha=0.6
)

plt.xticks(rotation=45, ha='right')
plt.yticks(
    ticks=range(len(jobsat_labels)),
    labels=jobsat_labels
)

plt.xlabel("Preferred Web Framework (WebframeWantToWorkWith)")
plt.ylabel("Job Satisfaction (JobSat)")
plt.title("Bubble Plot: Preferred Web Frameworks vs Job Satisfaction\n(Bubble size = number of respondents)")
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
# Keep only needed columns and drop missing
tmp = df[['Country', 'LanguageAdmired']].dropna()

# Split multi-select admired technologies and explode
tmp = tmp.assign(Tech=tmp['LanguageAdmired'].str.split(';'))
tmp = tmp.explode('Tech')

# Count respondents for each (Country, Tech) pair
admire = (
    tmp.groupby(['Country', 'Tech'])
       .size()
       .reset_index(name='count')
)

# Encode technologies for y-axis
admire['Tech_code'] = admire['Tech'].astype('category').cat.codes
tech_labels = admire['Tech'].astype('category').cat.categories

plt.figure(figsize=(12, 7))
plt.scatter(
    admire['Country'],       # x: country
    admire['Tech_code'],     # y: admired technology (coded)
    s=admire['count'] * 15,  # bubble size: frequency of admiration
    alpha=0.6
)

plt.xticks(rotation=90, ha='right')
plt.yticks(
    ticks=range(len(tech_labels)),
    labels=tech_labels
)

plt.xlabel("Country")
plt.ylabel("Admired Technology (LanguageAdmired)")
plt.title("Bubble Plot: Admired Technologies Across Countries\n(Bubble size = frequency of admiration)")
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
