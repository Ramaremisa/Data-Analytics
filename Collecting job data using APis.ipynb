```python
data = pd.read_csv("marketing_sample_jobs_data.csv")

def get_number_of_jobs_T(technology):
    # match technology anywhere in the Key Skills column, case-insensitive
    mask = data["Key Skills"].astype(str).str.contains(technology, case=False, na=False)
    number_of_jobs = int(mask.sum())
    return technology, number_of_jobs

print(get_number_of_jobs_T("Python"))
```

```python
def get_number_of_jobs_L(location):
    mask = data["Location"].astype(str).str.strip().str.casefold() == location.strip().casefold()
    number_of_jobs = int(mask.sum())
    return location, number_of_jobs

print(get_number_of_jobs_L("Mumbai"))
```

```python

technologies = ["Python", "Java", "Javascript", "C++", "AWS", "SQL", "Machine Learning", "Data Science"]
```

```python
from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title = "Data"

ws.title = "Job Postings"

# Write headers
ws.append(["Technology", "Number of Job Postings"])
```

```python
for tech in technologies:
    tech_name, num_jobs = get_number_of_jobs_T(tech)
    ws.append([tech_name, num_jobs])
```

```python
# --------------------------------------------
# Save into an Excel spreadsheet named job-postings.xlsx
# --------------------------------------------
wb.save("job-postings.xlsx")
print("✅ Excel file 'job-postings.xlsx' created successfully.")
```
